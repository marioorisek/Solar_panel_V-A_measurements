```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
coef = 0.85
```

```python
def calculate_current(light):
    return 3.88725e-6 * light

def calculate_voltage(light):
    if light > 0 and light <= 2000:
        return (3.06894274809285 * np.log(light)) - 7.76843407174411
    elif light > 2000 and light <= 5000:
        return (1.30084256686679 * np.log(light)) + 5.60633836103112
    elif light > 5000 and light < 140000:
        return (0.247002597209337 * np.log(light)) + 14.6442873809105
    else:
        return None
```

```python
svetlo = 70_000
print(f"Napeti: {calculate_voltage(svetlo):.1f} V Proud: {calculate_current(svetlo):.3f} A")
```

```python
svetlo = 110_000
print(f"Napeti: {calculate_voltage(svetlo):.1f} V Proud: {calculate_current(svetlo):.3f} A")

```

### Currenta

```python
light = range(0,120000, 10)
current = [3.88725e-6 * i for i in light]
current_minimal = [coef * i for i in current]
good = [500 for i in light]
bad = [0 for i in light]
```

```python
plt.figure(figsize=(20,10))
plt.text(22000, 0.3, "V podku", fontsize=48, fontweight='bold')
plt.text(85000, 0.12, "Vadn", fontsize=48, fontweight='bold')
plt.ylabel("Zkratov proud [A]", fontsize=24)
plt.xlabel("Intenzita osvtlen [lux]", fontsize=24)
plt.grid("both")
plt.plot(light, current_minimal, color='black', linewidth=5)
plt.fill_between(light, good, current_minimal, color='green', alpha=0.15)
plt.fill_between(light, bad, current_minimal, color='red', alpha=0.15)
plt.ylim((0, 0.5))
plt.xlim((0, 120_000))
plt.yticks(np.arange(start=0, stop=0.51, step=0.02))
plt.xticks(np.arange(start=0, stop=120_001, step=5000))
plt.show()
```

### Voltage

```python
light1 = range(10, 2_000, 10)
voltage1 = [coef * ((3.06894274809285 * np.log(i)) - 7.76843407174411) for i in light1]
good1 = [20 for i in light1]
bad1= [0 for i in light1]

light2 = range(2_000,5_000, 10)
voltage2 = [coef * ((1.30084256686679 * np.log(i)) + 5.60633836103112) for i in light2]
good2 = [20 for i in light2]
bad2 = [0 for i in light2]

light3 = range(5_000,125_001, 10)
voltage3 = [coef * ((0.247002597209337 * np.log(i)) + 14.6442873809105) for i in light3]
good3 = [20 for i in light3]
bad3 = [0 for i in light3]

```

```python
plt.figure(figsize=(20,10))
plt.text(22000, 17, "V podku", fontsize=48, fontweight='bold')
plt.text(65000, 8, "Vadn", fontsize=48, fontweight='bold')
plt.ylabel("Napt na przdno [V]", fontsize=24)
plt.xlabel("Intenzita osvtlen [lux]", fontsize=24)
plt.grid("both")

plt.plot(light1, voltage1, color='black', linewidth=5)
plt.fill_between(light1, good1, voltage1, color='green', alpha=0.15)
plt.fill_between(light1, bad1, voltage1, color='red', alpha=0.15)

plt.plot(light2, voltage2, color='black', linewidth=5)
plt.fill_between(light2, good2, voltage2, color='green', alpha=0.15)
plt.fill_between(light2, bad2, voltage2, color='red', alpha=0.15)

plt.plot(light3, voltage3, color='black', linewidth=5)
plt.fill_between(light3, good3, voltage3, color='green', alpha=0.15)
plt.fill_between(light3, bad3, voltage3, color='red', alpha=0.15)

plt.yticks(np.arange(start=0, stop=20.1, step=1))
plt.xticks(np.arange(start=0, stop=120_001, step=5000))

plt.ylim((0, 20.1))
plt.xlim((0, 120_800))

plt.show()
```
